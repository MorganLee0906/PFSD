```python
import os
import glob


def convert_sas_to_csv(sas_file):
    with open(sas_file, 'r', encoding='BIG5', errors='ignore') as f:
        lines = f.readlines()
        
        # 替換每一行中的逗號為中文逗號
    for i in range(len(lines)):
        lines[i] = lines[i].replace(",", "，")

    csv_answers = []
    csv_options = []
    i = 0
    while not lines[i+1].startswith('format') and not lines[i+1].startswith('FORMAT') and not lines[i+1].startswith('   FORMAT'):
        if lines[i].startswith('   value') or lines[i].startswith('   VALUE'):
            answer = lines[i].split()[1]
            j = i + 1
            while not lines[j+1].startswith('format') and not lines[j+1].startswith('FORMAT') and not lines[j+1].startswith('   FORMAT') and not lines[j].startswith('   value') and not lines[j].startswith('   VALUE'):
                csv_answers.append(answer)
                csv_options.append(lines[j].strip())
                #print(sas_file, answer, lines[j].strip())
                j += 1
        i += 1

    dir_name = os.path.basename(os.path.dirname(sas_file))
    csv_file = os.path.join(os.path.dirname(sas_file), dir_name + '_answer.csv')
    
    with open(csv_file, 'w', encoding='utf-8') as f:
        f.write("YEAR,ANSWER,OPTION\n")
        for n in range(len(csv_answers)):
            f.write(dir_name + ',' + csv_answers[n] + ',' + csv_options[n] + '\n')

if __name__ == '__main__':
    data_folder = '/Users/virginialiu/Desktop/cclab/PSFD/data/子女樣本'
    for subdir in os.listdir(data_folder):
        subdir_path = os.path.join(data_folder, subdir)
        if os.path.isdir(subdir_path):
            sas_files = glob.glob(os.path.join(subdir_path, '*.sas'))
            for sas_file in sas_files:
                print(sas_file)
                convert_sas_to_csv(sas_file)
                print(subdir)
```
